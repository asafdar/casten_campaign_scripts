```python
import pandas as pd

df_master = pd.read_excel(r'myc.xlsx')
df_master.head()
```

```python
df_mini = pd.read_csv(r'bsd.csv')
df_mini.head()
```

```python
new_df = pd.DataFrame()
```

```python
new_df['Is Duplicate?'] = ''
```

```python
df_mini = df_mini.rename({'firstname':'FirstName','lastname':'LastName'}, axis='columns') 
```

```python
import numpy as np

duplicate_indices = df_mini.merge(df_master, on=['FirstName', 'LastName'], how='inner').index
df_mini['isDuplicate'] = np.where(df_mini.index.isin(duplicate_indices), 'Yes', 'No')
```

```python
df_mini.head()
```

```python
df_mini.to_csv('myc_duplicates.csv')
```

```python
for person in df_mini:
    if (df_mini.FirstName == df_master.FirstName) and (df_mini.LastName == df_master.LastName):
        new_df['Is Duplicate?'].append('Yes')
    else:
        new_df['Is Duplicate?'].append('No')
```
