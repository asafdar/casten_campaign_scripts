```python
import json
import requests
import time
import urllib.parse
import pandas as pd
```

```python
api_token="AIzaSyByRSmKNiKpay7rSyvbyq1G7HiXBh7GvRo"
api_url_base = "https://www.googleapis.com/civicinfo/v2/representatives"
```

```python
headers = {'Content-Type': 'application/json'}
```

```python
def get_district_info(address):
    #Build the URL, inserting the address, key, and asking for just the divisions back.
    api_url = '{0}?address={1}&key={2}&fields=divisions'.format(api_url_base,urllib.parse.quote(address),api_token)
    #Make the call.
    response = requests.get(api_url, headers=headers)
    #If the server returned successfully, return what it gave us, otherwise return None.
    if response.status_code == 200:
        return json.loads(response.content.decode('utf-8'))
    else:
        return None
```

```python
#Load the addresses.
addressFile = open("addresses.csv","r")
```

```python
cols = ['Address', 'District']
df = pd.DataFrame(columns = cols)
```

```python
for address in addressFile:
  #Strip off the newline from the address
  address=address.rstrip()
  #Perform the call to the API with the address.
  result = get_district_info(address)
  district = "Unknown"
  if result == None:
    district = "Lookup Failed"
  else:
    #Iterate over all of the divisions we got back.
    for division in result["divisions"]:
        #Pick out just the congressional district division.
        if "/cd:" in division:
          #Success!  We set district to what we found.
          district=result["divisions"][division]["name"]
  #Append results to dataframe
  df = df.append({'Address': address, 'District': district}, ignore_index=True)
  #Output what we found
  print('"',address,'","',district,'"',sep="")
  #Wait three quarters of a second so as not to make too many requests too quickly.
  time.sleep(.5)
```
